```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def power_function(samp_size, thresh):
    params = np.arange(0,2.02,.02)
    results = np.array([])
    for param in params:
        results = np.append(results,1 - (1-np.e**(-param*thresh))**samp_size)
    return results

samp_sizes = range(200,2200,400)
threshes = range(7,11)
                
plt.figure(figsize=(12,12))
for i in range(1,len(threshes)+1):
    plt.subplot(len(threshes),1,i)
    for samp_size in samp_sizes:
        lab = "N = " + str(samp_size)
        title = "Power function for threshold = " + str(threshes[i-1])
        plt.plot(np.arange(0,2.02,.02), power_function(samp_size, threshes[i-1]),label=lab)
        plt.axvline(1, c='k',ls='--')
        plt.axhline(0.05, c='k',ls='--')
        plt.title(title)
        plt.legend(loc=7)
    plt.xlabel(r"\lambda")
    plt.ylabel("Power")
    
q2c_fig = plt.gcf()
q2c_fig.savefig('Q2c_fig.png', bbox_inches='tight')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os.path

plt.close("all")
men_chol = np.loadtxt("cholesterol_men.txt", delimiter=',') 
women_chol = np.loadtxt("cholesterol_women.txt", delimiter=',') 

def run_permutation_test_mean(data, n1, n2, n = 100000):
    res_mean = np.array([])
    for i in xrange(n):
        men_index = np.random.choice(np.arange(len(data)), size=n1, replace=False)
        women_index = np.setdiff1d(np.arange(len(data)), men_index)
        men_sample = data[men_index]
        women_sample = data[women_index]
        mean_diff = men_sample.mean() - women_sample.mean()
        res_mean = np.append(res_mean, mean_diff)
    return res_mean
    
def run_permutation_test_median(data, n1, n2, n = 100000):
    res_mean = np.array([])
    for i in xrange(n):
        men_index = np.random.choice(np.arange(len(data)), size=n1, replace=False)
        women_index = np.setdiff1d(np.arange(len(data)), men_index)
        men_sample = data[men_index]
        women_sample = data[women_index]
        mean_diff = np.median(men_sample) - np.median(women_sample)
        res_mean = np.append(res_mean, mean_diff)
    return res_mean

def compute_p_value(res, val):
    extreme_vals = res[(res > val)]
    return float(len(extreme_vals))/float(len(res))
    

test_statistic_mean = men_chol.mean() - women_chol.mean()
test_statistic_median = np.median(men_chol) - np.median(women_chol)
data = np.append(men_chol, women_chol)

print "Difference in sample mean is " + str(test_statistic_mean)
print "Difference in sample median is " + str(test_statistic_median)

tries = 3
n = 100000
for i_try in range(tries):
    res_mean = run_permutation_test_mean(data, len(men_chol), len(women_chol))
    res_median = run_permutation_test_median(data, len(men_chol), len(women_chol))
    p_value_mean = compute_p_value(res_mean, test_statistic_mean)
    p_value_median = compute_p_value(res_median, test_statistic_median)
    print "p value for difference in sample mean is " + str(p_value_mean)
    print "p value for difference in sample median is " + str(p_value_median)
    plt.figure(figsize=(12, 9))  
    ax = plt.subplot(111)    
    ax.spines["top"].set_visible(False)    
    ax.spines["right"].set_visible(False)   
    ax.get_xaxis().tick_bottom()  
    ax.get_yaxis().tick_left()  
    plt.xticks(fontsize=22) 
    plt.yticks(fontsize=22) 
    plt.xlim([-25,25])
    plt.ylim([0,0.075])
    ax.set_xticks([-20, -10, 0, 10, test_statistic_mean])
    # plt.xlabel("Mean(men) - mean(woem)", fontsize=20)      
    plt.hist(res_mean,100,normed=True,edgecolor = "none", color="skyblue", label='Approximate pdf under null hypothesis')
    plt.plot([test_statistic_mean, test_statistic_mean],[0, 0.075],ls="dashed", color="darkred", lw=2)
    plt.title("Difference in sample mean")
    
    plt.figure(figsize=(12, 9))  
    ax = plt.subplot(111)    
    ax.spines["top"].set_visible(False)    
    ax.spines["right"].set_visible(False)   
    ax.get_xaxis().tick_bottom()  
    ax.get_yaxis().tick_left()  
    plt.xticks(fontsize=22) 
    plt.yticks(fontsize=22) 
    plt.xlim([-25,25])
    #plt.ylim([0,0.075])
    ax.set_xticks([-20, -10, 0, 10, test_statistic_median])
    # plt.xlabel("Mean(men) - mean(woem)", fontsize=20)      
    plt.hist(res_median,100,normed=True,edgecolor = "none", color="skyblue", label='Approximate pdf under null hypothesis')
    plt.plot([test_statistic_median, test_statistic_median],[0, 0.11],ls="dashed", color="darkred", lw=2)    
    plt.title("Difference in sample median")
```
