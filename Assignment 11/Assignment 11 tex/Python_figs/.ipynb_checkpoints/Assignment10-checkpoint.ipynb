```python
import os.path
import math
import matplotlib.pyplot as plt
import numpy as np

plt.close("all")
plt.rcParams['xtick.major.pad']='8'
plt.rcParams['ytick.major.pad']='8'

### CHANGE PATH TO WHEREVER YOU SAVE THE DATA FILES ### 
dataset = np.loadtxt("oxford_temperatures.txt")
max_temp = np.array(dataset[:,2])
min_temp = np.array(dataset[:,3])
year = np.array(dataset[:,0])
n = len(max_temp)
```

```python
# First build matrix A:
import pandas as pd

t = np.arange(n)
A = pd.concat([pd.Series(np.ones(n).T, name='One'),\
               pd.Series(np.cos(2.0*np.pi*t/12.0).T, name='Cos'),\
               pd.Series(np.sin(2.0*np.pi*t/12.0).T, name='Sin'),\
               pd.Series(t.T, name='Month')], axis=1)
```

```python
### INSERT CODE HERE ### 
# Suggestion: Use np.linalg.lstsq to solve the least-squares problem.


#Then fit models
max_model_coef = np.linalg.lstsq(A.values, max_temp)[0]
def max_model(t):
    y_fit = max_model_coef[0] +\
    max_model_coef[1]*np.cos(2.0*np.pi*t/12.0) +\
    max_model_coef[2]*np.sin(2.0*np.pi*t/12.0) +\
    max_model_coef[3]*t
    return y_fit

def max_trend(t):
    y_fit = max_model_coef[0] + max_model_coef[3]*t
    return y_fit

min_model_coef = np.linalg.lstsq(A.values, min_temp)[0]
def min_model(t):
    y_fit = min_model_coef[0] +\
    min_model_coef[1]*np.cos(2.0*np.pi*t/12.0) +\
    min_model_coef[2]*np.sin(2.0*np.pi*t/12.0) +\
    min_model_coef[3]*t
    return y_fit

def min_trend(t):
    y_fit = min_model_coef[0] + min_model_coef[3]*t
    return y_fit

reconstruction_max = max_model(t)
reconstruction_min = min_model(t)
trend_max =  max_trend(t)
trend_min =  min_trend(t)
```

```python
#No intercept needed

eps = np.random.randn(1000)
xs = np.random.random(1000)*10
ys = xs + eps

fig = plt.figure(figsize=(10,5))

##No intercept needed
ax1 = plt.subplot(121)
ax1.scatter(xs, ys)
ax1.set_xlim(0,10)
line = np.polyfit(xs,ys,1)
x_for_line = np.linspace(0,10,2)
y_for_line = line[0]*x_for_line + line[1]
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title(r'Relationship in $\mathbb{R}^2$ with no intercept needed')
ax1.plot(x_for_line, y_for_line, 'k-', linewidth=8, label='Regression fit')
ax1.legend()

##Intercept needed
ys_new = ys  + 6
ax2 = plt.subplot(122)
ax2.scatter(xs, ys_new)
ax2.set_xlim(0,10)
ax2.set_ylim(0,18)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
line2 = np.polyfit(xs,ys_new,1)
x_for_line2 = np.linspace(0,10,2)
y_for_line2 = line2[0]*x_for_line2 + line2[1]

fit_no_int = np.linalg.lstsq(xs[:, np.newaxis], ys_new)[0]
y_no_int = fit_no_int*x_for_line2


ax2.set_title(r'Relationship in $\mathbb{R}^2$ with intercept needed')
ax2.plot(x_for_line2, y_for_line2, 'k-', linewidth=8, label='Regression fit with intercept')
ax2.plot(x_for_line2, y_no_int, 'g-', linewidth=8, label='Regression fit without intercept')
ax2.legend(loc='best')
```

```python
"""
"""
import os.path
import math
import matplotlib.pyplot as plt
import numpy as np

plt.close("all")
plt.rcParams['xtick.major.pad']='8'
plt.rcParams['ytick.major.pad']='8'

### CHANGE PATH TO WHEREVER YOU SAVE THE DATA FILES ### 
data = np.loadtxt("heights_weights.txt") 

n_train = 20000
n_test = 1000

# Randomly select training, test and validation sets
np.random.shuffle(data)
data_train = data[range(n_train),:]
data_test = data[range(n_train, n_train+n_test),:]
heights_train = data_train[:,1]
weights_train = data_train[:,2]
heights_test = data_test[:,1]
weights_test = data_test[:,2]
```

```python
### INSERT CODE HERE ###
# Hint: To fit model 1 use the expression you found directly
# to fit model 2 you can use np.linalg.lstsq

#Fit no intercept
alpha = np.dot(np.dot(1.0/np.dot(weights_train.T, weights_train),weights_train.T), heights_train)
pred_model_1 = alpha*weights_test

#Fit with intercept
A = np.hstack([weights_train[:,np.newaxis], np.ones_like(weights_train)[:,np.newaxis]])
coefs, resids, rank, singular = np.linalg.lstsq(A, heights_train)
pred_model_2 = coefs[0]*weights_test + coefs[1]*np.ones_like(weights_test)

plt.figure(figsize=(12, 9))  
plt.plot(weights_test, heights_test, '.',  color='skyblue',markeredgecolor='blue')
plt.plot(weights_test, pred_model_1,  c="red")

plt.figure(figsize=(12, 9))  
plt.plot(weights_test, heights_test, '.', color='skyblue',markeredgecolor='blue')
plt.plot(weights_test, pred_model_2, c="red")

error_1 = np.sum(np.abs((pred_model_1 - heights_test)/heights_test))/n_test
error_2 = np.sum(np.abs((pred_model_2 - heights_test)/heights_test))/n_test

print "Model 1 relative error: " + str(error_1)
print "Model 2 relative error: " + str(error_2)
```

Now we repeat the fitting process after downsampling our training data:

## Note this is for model2 with the intercept:

```python
def sample_size_error_test(size):
    n_train = size
    n_test = (21000 - n_train)

    # Randomly select training, test and validation sets
    np.random.shuffle(data)
    data_train = data[range(n_train),:]
    data_test = data[range(n_train, n_train+n_test),:]
    heights_train = data_train[:,1]
    weights_train = data_train[:,2]
    heights_test = data_test[:,1]
    weights_test = data_test[:,2]

    #Fit with intercept
    A = np.hstack([weights_train[:,np.newaxis], np.ones_like(weights_train)[:,np.newaxis]])
    coefs, resids, rank, singular = np.linalg.lstsq(A, heights_train)
    pred_model_2 = coefs[0]*weights_test + coefs[1]*np.ones_like(weights_test)

    error_2 = float(np.sum(np.abs((pred_model_2 - heights_test)/heights_test))/n_test)

    return error_2
```

```python
def many_trials_make_graph(num_trials):
    results = {}
    for trial in range(num_trials):
        for samp_size in range(2,100,1):
            if samp_size in results.keys():
                results[samp_size].append(sample_size_error_test(samp_size))
            else:
                results[samp_size] = [sample_size_error_test(samp_size)]
    return results
```

```python
results1 = many_trials_make_graph(10)
```

```python
def sample_size_error_test_no_int(size):
    n_train = size
    n_test = (21000 - n_train)

    # Randomly select training, test and validation sets
    np.random.shuffle(data)
    data_train = data[range(n_train),:]
    data_test = data[range(n_train, n_train+n_test),:]
    heights_train = data_train[:,1]
    weights_train = data_train[:,2]
    heights_test = data_test[:,1]
    weights_test = data_test[:,2]

    #Fit with intercept
    
    alpha = np.dot(np.dot(1.0/np.dot(weights_train.T, weights_train),weights_train.T), heights_train)
    pred_model_1 = alpha*weights_test
    error_2 = float(np.sum(np.abs((pred_model_1 - heights_test)/heights_test))/n_test)

    return error_2
```

```python
def many_trials_make_graph_no_int(num_trials):
    results = {}
    for trial in range(num_trials):
        for samp_size in range(2,100,1):
            if samp_size in results.keys():
                results[samp_size].append(sample_size_error_test_no_int(samp_size))
            else:
                results[samp_size] = [sample_size_error_test_no_int(samp_size)]
    return results
```

```python
results2 = many_trials_make_graph_no_int(10)
```

```python
df = pd.DataFrame(results1)
df2 = pd.DataFrame(results2)
fig2 = plt.figure(figsize=(10,5))

ax2 = plt.subplot(121)
ax2.plot(df2.mean(), label='Mean relative error')
ax2.plot(df2.mean()+df2.std()/10.0, 'k--', label='Mean +/- std/n')
ax2.plot(df2.mean()-df2.std()/10.0, 'k--')
ax2.legend(loc='best')
ax2.set_xlabel('Sample size (n)')
ax2.set_xlabel('Relative error')
ax2.set_title('Model 1: Without Intercept')


ax1 = plt.subplot(122)
ax1.plot(df.mean(), label='Mean relative error')
ax1.plot(df.mean()+df.std()/10.0, 'k--', label='Mean +/- std/n')
ax1.plot(df.mean()-df.std()/10.0, 'k--')
ax1.legend(loc='best')
ax1.set_xlabel('Sample size (n)')
ax1.set_xlabel('Relative error')
ax1.set_title('Model 2: With Intercept')
```
